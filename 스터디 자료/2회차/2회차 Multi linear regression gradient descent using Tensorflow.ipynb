```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import random

data = [[2,0,81],[4,4,93],[6,2,91],[8,3,97]]
x1=[i[0] for i in data]
x2=[i[1] for i in data]
y=[i[2] for i in data]
```

```python
ax = plt.axes(projection='3d')
ax.set_xlabel('study hours')
ax.set_ylabel('private_class')
ax.set_zlabel('Score')
ax.dist=11
ax.scatter(x1,x2,y)
plt.show()
```

```python
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)
```

```python

a1 = tf.Variable(random.random())
a2 = tf.Variable(random.random())
b = tf.Variable(random.random())

def compute_loss():
    y_pred = a1 * x1_data + a2 * x2_data + b
    loss = tf.reduce_mean((y-y_pred)**2)
    return loss

optimizer = tf.optimizers.Adam(lr=0.1)
epoch = 5001

for i in range(epoch):
    optimizer.minimize(compute_loss, var_list=[a1,a2,b])
    
    if i%500==0:
        print("epoch=",i,"기울기1=",a1.numpy(),"기울기2=",a2.numpy(),"절편=",b.numpy(),"loss=",compute_loss().numpy())
```

```python
y_pred = a1 * x1_data + a2 * x2_data + b
ax = plt.axes(projection='3d')
ax.scatter(x1,x2,y)
plt.plot([min(x1_data),max(x1_data)], [min(x2_data),max(x2_data)],[min(y_pred),max(y_pred)])
plt.show()
```
